```python
# Get the dataframe of district and ward name
df_dn_district = pd.read_excel("danang_street_name.xlsx", sheet_name="district, ward")

# Rename
df_dn_district = df_dn_district.rename(
    columns={"Quận Huyện": "district", "Phường Xã": "ward"}
)

# Clean for district column
for i in range(len(df_dn_district['district'])):
  try:
    df_dn_district['district'][i] = re.sub('[^A-Za-z0-9]+', ' ', (unidecode(df_dn_district['district'][i])).lower())
  except:
    df_dn_district['district'][i] = 0

# Clean for ward column
for i in range(len(df_dn_district['ward'])):
  try:
    df_dn_district['ward'][i] = re.sub('[^A-Za-z0-9]+', ' ', (unidecode(df_dn_district['ward'][i])).lower())
  except:
    df_dn_district['ward'][i] = df_dn_district['district'][i]

df_dn_district
```

```python
# Now, get the list of district name and strim a part of unuse information
df_dn_district["district_new"] = df_dn_district["district"].map(
    lambda x: x.lstrip("quan|huyen")
)
df_dn_district["district_new"] = df_dn_district["district_new"].str.lstrip(" ")
# Also do it with ward
df_dn_district["ward_new"] = df_dn_district["ward"].map(
    lambda x: x.lstrip("huyen|phuong|xa")
)
df_dn_district["ward_new"] = df_dn_district["ward_new"].str.lstrip(" ")
# Get a list of unique district and ward
list_dn_district = list(df_dn_district["district_new"].unique())
list_dn_ward = list(df_dn_district["ward_new"].unique())
list_dn_ward.append('hoa khanh')
```

```python
# Get a list of streetname
df_dn_street['street_name'] = df_dn_street['street_name'].str.lstrip(" ")
df_dn_street['street_name'] = df_dn_street['street_name'].str.rstrip(" ")
list_dn_street = list(df_dn_street['street_name'].unique())
```

```python
# The first code line is to check whether the address contains "dn" or "da nang"
df_customer['contains_dn'] = df_customer['address'].str.contains(('dn|da nang'))

# Then test with district list and ward list
df_customer['contains_dt'] = df_customer['address'].str.contains('|'.join(list_dn_district))
df_customer['contains_w'] = df_customer['address'].str.contains('|'.join(list_dn_ward))

# Finally, test with streetname
df_customer['contains_str'] = df_customer['address'].str.contains('|'.join(list_dn_street))
```

There's about 96 data in the ticket["orderid"] are NaN/NA value --> Order from web
Solution: Assign new orderid for web.
There're some tendency of people orderred from website:
- Same cashier (which is Website)
- Same customerid
- Same film
- Same date
- Same time
- Same room

```python
# Get the different of saledate and showdate
df_ticket.loc[df_ticket["cashier"]!="Website", "saledate-showdate"] = df_ticket["showdate"]-df_ticket["saledate"]
df_ticket["saledate-showdate"] = df_ticket["saledate-showdate"].dt.days
# Calculate the average difference
print("Thời gian trung bình từ ngày đặt vé đến ngày xem phim: ", round(df_ticket["saledate-showdate"].mean()), " ngày")
```

```python
# Append the average date different to cashier is website
df_ticket.loc[df_ticket["cashier"]=="Website", "saledate-showdate"] = round(df_ticket["saledate-showdate"].mean())
```

```python
df_ticket_1 = df_ticket.copy()
```

```python
# Assign the order id for web order
# Giả sử, với same customerid và same showdate, người mua vé mua cùng một orderid
df_web_1 = df_ticket_1[["customerid", "showdate"]]
df_web_1
```

```python
# Get the new orderid for people order on website
df_ticket_1.loc[(df_ticket_1["cashier"] == "Website") & (df_ticket_1["orderid"].isnull()), "orderid"] = "website" + df_ticket_1["customerid"].astype(str) + df_ticket_1["showdate"].astype(str)

# Assume saledate is 15 days before the showdate
# df_ticket_1["saledate"] = df_ticket_1["showdate"] - pd.to_timedelta(15, unit = "D")
df_ticket_1["saledate"] = df_ticket_1["showdate"]
```

```python
# Append the No popcorn to cashier is website
df_ticket_1.loc[df_ticket_1["cashier"]=="Website", "popcorn"] = "Không"
```
